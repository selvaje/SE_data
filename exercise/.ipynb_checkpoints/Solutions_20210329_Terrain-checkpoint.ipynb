## Q1 (Bonus, 10%) : 

Represent the Eucledian distance formula using the spherical coordinates. 

$$d(A,B) = \sqrt{(A_x - B_x)^2 + (A_y - B_y)^2 + (A_z - B_z)^2}$$


coordinates conversion : catesian to spherical 

$$
\left\{
\begin{aligned}
z &= r\, cos(\theta) \\
x &= r\, sin(\theta)\, cos(\phi) \\
y &= r\, sin(\theta)\, sin(\phi)
\end{aligned}
\right.
$$


## Solution : QB1

\begin{align*}
d(A,B) &= \sqrt{(r_A sin\theta_A cos\phi_A - r_B sin\theta_B cos\phi_B)^2 + (r_A sin\theta_A sin\phi_A - r_B sin\theta_B sin\phi_B)^2 + (r_A cos\theta_A - r_B cos\theta_B)^2}\\
       &= \sqrt{r_A sin^2\theta_A cos^2\phi_A + r_B sin^2\theta_B cos^2\phi_B - 2r_Ar_B sin\theta_A cos\phi_A sin\theta_B cos\phi_B +r_A sin^2\theta_A sin^2\phi_A + r_B sin^2\theta_B sin^2\phi_B - 2r_Ar_B sin\theta_A sin\phi_A sin\theta_B sin\phi_B + r^2_Acos^2\theta_A + r^2_Bcos^2\theta_B - 2r_Ar_B cos\theta_A cos\theta_B} \\
       &= \sqrt{r^2_A sin^2\theta_A + r^2_B sin^2\theta_B + r^2_A cos^2\theta_A + r^2_B cos^2\theta_B - 2r_Ar_B (sin\theta_A cos\phi_A sin\theta_B cos\phi_B + sin\theta_A sin\phi_B sin\theta_B sin\phi_B + cos\theta_A cos\theta_B)}\\
       &= \sqrt{r^2_A + r^2_B - 2r_Ar_B(sin\theta_A sin\theta_B(cos(\phi_A - \phi_B))+ cos\theta_A cos\theta_B)}\\
       &= \sqrt{r^2_A + r^2_B - r_Ar_B((cos(\theta_A - \theta_B) - cos(\theta_A + \theta_B))(cos(\phi_A - \phi_B) + cos(\theta_A - \theta_B) + cos(\theta_A + \theta_B))}      
\end{align*}


## Q2 (Bonus, 10%) : 

Atmospheric pressure $P(z)$(hPa) as a function of altitude z (km) in the troposphere

$$P(z) = 1013.25\, \left(\dfrac{288 - 6.5z}{288}\right)^{5.255}$$

Compute the gradient of $P(z)$ and plot it as a function of $z$ in the range ($1 < z <12$)

## Solution : QB2 

```r
P <- expression(1013.25*((288-6.5*z)/288)^5.255)
dPdz <- D(P,"z")
dPdz
```

```r
z <- seq(1,12,length.out=100)
plot(z,eval(dPdz),type="l",lwd=3,cex.lab=1.5,cex.axis=1.5,ylab="dP/dz",xlab="z",xlim=c(1,12))
grid(lwd=3,nx=12)
```
