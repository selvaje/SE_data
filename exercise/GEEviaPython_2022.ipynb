### Install Google stuff for authentication with 2FA and gcloud

See <https://cloud.google.com/sdk/docs/install-sdk#deb> for details and multiple distributions and platforms.

In brief for OSGeoLive/Debian/Ubuntu distros:

```
sudo apt-get install apt-transport-https ca-certificates gnupg
```

```
echo "deb [signed-by=/usr/share/keyrings/cloud.google.gpg] https://packages.cloud.google.com/apt cloud-sdk main" | \
```

```
sudo tee -a /etc/apt/sources.list.d/google-cloud-sdk.list 
```

```
wget -O- https://packages.cloud.google.com/apt/doc/apt-key.gpg | sudo tee /usr/share/keyrings/cloud.google.gpg >/dev/null
```

```
sudo apt-get update && sudo apt-get install google-cloud-cli
```
